```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('598354.csv', infer_datetime_format=True, parse_dates=["DATE"])
d1 = df[df["TMAX"] != -9999]
d2 = d1[d1["TMIN"] != -9999]
d3 = d2[d2["PRCP"] != -9999]
stations = d3.groupby("STATION")
station = stations.get_group(stations.groups.keys()[1])
```

```python
plt.figure(1, figsize=(15,8))
plt.title('Temperature in ' + str(station.STATION_NAME.unique()[0]))
plt.xlabel('Dates')
plt.ylabel('Temperature $(^{\circ{c}})$')
plt.plot(station.DATE, station.TMAX / 10, 'r-')
plt.plot(station.DATE, station.TMIN / 10, 'b-')
```

```python
plt.figure(2, figsize=(15,8))
plt.title('Precipitation in ' + str(station.STATION_NAME.unique()[0]))
plt.xlabel('Dates')
plt.ylabel('Precipitation $mm \cdot 10^{-1}$')
date_indexed = station.set_index(station.DATE)
plt.plot(date_indexed.DATE['2015-6'], date_indexed.PRCP['2015-6'], 'b-')
```
